```python
x= df["Rating"]
y= df["Reviews"]
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y, c="k", marker = "*", s= 100, alpha= 0.5, linewidths=5, edgecolors="b")#verts="<" 
plt.title("Google Play Store Apps Scatter Plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```
