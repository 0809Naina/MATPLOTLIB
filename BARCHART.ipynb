```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
classes =["Py","R","AI","ML","DS"]
class1_studs =[30,10,20,25,10]
class2_studs =[40,5,20,20,10]
class3_studs =[35,5,30,15,15]
```

```python
plt.bar(classes, class1_studs)
```

```python
plt.barh(classes, class1_studs)
```

```python
plt.bar(classes, class1_studs,width=0.2)
```

```python
plt.bar(classes, class1_studs,width=0.2,align="edge",color="b",edgecolor="k", linewidth=5,alpha=0.4,linestyle=":", label="Class 1 Students")#visible=False 
```

```python
#alpha is used for the the colour shade range.
```

```python
style.use("ggplot")
plt.figure(figsize=(16,8))
plt.bar(classes, class1_studs,width=0.2,align="edge",color="b",edgecolor="k", linewidth=5,alpha=0.4,linestyle=":", label="Class 1 Students")#visible=False 
```

```python
style.use("ggplot")
plt.figure(figsize=(16,8))
plt.barh(classes, class1_studs,align="edge",color="b",edgecolor="k", linewidth=5,alpha=0.4,linestyle=":", label="Class 1 Students")#visible=False 
```

```python
style.use("ggplot")
plt.figure(figsize=(16,8))
plt.bar(classes, class1_studs,width=0.2,align="edge",color="b",edgecolor="k", linewidth=5,alpha=0.4,linestyle=":", label="Class 1 Students")#visible=False 
plt.title("Bar Chart of IAIP Class", fontsize= 19)
plt.xlabel("classes", fontsize= 15)
plt.ylabel("No. of Students",fontsize= 15)
plt.show() 
```

```python
style.use("ggplot")
plt.figure(figsize=(16,8))

classes_index = np.arange(len(classes))

width=0.2

plt.bar(classes_index, class1_studs,width=0.2,color="k",label="Class 1 Students")#visible=False

plt.bar(classes_index + width, class2_studs,width=0.2,color="grey",label="Class 2 Students")

plt.bar(classes_index + width +width , class3_studs,width=0.2,color="b", label="Class 3 Students")

plt.xticks(classes_index+ width,classes , rotation =20) 
plt.title("Bar Chart of IAIP Class", fontsize= 19)
plt.xlabel("classes", fontsize= 19)
plt.ylabel("No. of Students",fontsize= 19)
plt.show()
```

```python
style.use("ggplot")
plt.figure(figsize=(16,8))

classes_index = np.arange(len(classes))

width=0.2

plt.barh(classes_index, class1_studs,width,color="k",label="Class 1 Students")#visible=False

plt.barh(classes_index + width, class2_studs,width,color="grey",label="Class 2 Students")

plt.barh(classes_index + width +width , class3_studs,width,color="b", label="Class 3 Students")

plt.xticks(classes_index+ width,classes , rotation =20) 
plt.title("Bar Chart of IAIP Class", fontsize= 19)
plt.xlabel("classes", fontsize= 19)
plt.ylabel("No. of Students",fontsize= 19)
plt.legend()
plt.show()
```
