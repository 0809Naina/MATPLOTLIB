```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
ml_students_age= np.random.randint(18,45,(100))
py_students_age= np.random.randint(15,40,(100))
```

```python
print(ml_students_age)
print(py_students_age)
```

```python
plt.hist (ml_students_age)

plt.title("ML STUDENTS AGE HISTOGRAM")
plt.xlabel("students age category")
plt.ylabel("No. of students age")
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist (ml_students_age,bins, rwidth=0.8, histtype="step")

plt.title("ML STUDENTS AGE HISTOGRAM")
plt.xlabel("students age category")
plt.ylabel("No. of students age")
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist (ml_students_age,bins, rwidth=0.8, histtype="bar")

plt.title("ML STUDENTS AGE HISTOGRAM")
plt.xlabel("students age category")
plt.ylabel("No. of students age")
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist (ml_students_age,bins, rwidth=0.8, histtype="bar", orientation="vertical", color="b", label="ML STUDENT")
plt.hist (py_students_age,bins, rwidth=0.8, histtype="bar", orientation="vertical", color="k", label="PY STUDENT")

plt.title("ML STUDENTS AGE HISTOGRAM")
plt.xlabel("students age category")
plt.ylabel("No. of students age")
plt.legend()
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist ([ml_students_age,py_students_age],bins, rwidth=0.8, histtype="bar", orientation="vertical", color=["b","k"], label=["ML STUDENT","PY STUDENT"])
#plt.hist (py_students_age,bins, rwidth=0.8, histtype="bar", orientation="vertical", color="k", label="PY STUDENT")

plt.title("ML STUDENTS AGE HISTOGRAM")
plt.xlabel("students age category")
plt.ylabel("No. of students age")
plt.legend()
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]

plt.figure(figsize=(16,8))
plt.hist ([ml_students_age,py_students_age],bins, rwidth=0.8, histtype="bar", orientation="vertical", color=["b","k"], label=["ML STUDENT","PY STUDENT"])
#plt.hist (py_students_age,bins, rwidth=0.8, histtype="bar", orientation="vertical", color="k", label="PY STUDENT")

plt.title("ML STUDENTS AGE HISTOGRAM")
plt.xlabel("students age category")
plt.ylabel("No. of students age")
plt.legend()
plt.show()
```

```python
from matplotlib import style 
style.use("ggplot")
plt.figure(figsize=(16,8))
plt.hist ([ml_students_age,py_students_age],bins, rwidth=0.8, histtype="bar", orientation="vertical", color=["b","k"], label=["ML STUDENT","PY STUDENT"])
#plt.hist (py_students_age,bins, rwidth=0.8, histtype="bar", orientation="vertical", color="k", label="PY STUDENT")

plt.title("ML STUDENTS AGE HISTOGRAM")
plt.xlabel("students age category")
plt.ylabel("No. of students age")
plt.legend()
plt.grid(color="w", linestyle='-.', linewidth=2)
plt.show()
```
