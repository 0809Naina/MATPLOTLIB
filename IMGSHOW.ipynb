# HOW TO SHOW IMAGE USING MATPLOTLIB IMSHOW FUNCTION?

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img =mpimg.imread("pie_char.png")
```

```python
plt.imshow(img)
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.axis("off")
plt.imshow(img)
plt.colorbar()
plt.show()
```

```python
 plt.figure(figsize=(16,9))
plt.axis("off")
plt.imshow(img, cmap="Accent")
plt.colorbar()
plt.show()
```

```python
single_channel= img[:,:,0]
plt.figure(figsize=(16,9))
plt.axis("off")
plt.imshow(img, cmap="hot")
plt.colorbar()
plt.show()
```
