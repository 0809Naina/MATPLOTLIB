```python
classes =["Py","R","AI","ML","DS"]
class1_studs =[45,15,35,25,30]
#class2_studs =[40,5,20,20,10]
#class3_studs =[35,5,30,15,15]
explode=[0.03,0,0.1,0,0]
colors= ["c","b","r","y","g"]
textprops={"fontsize":14}


plt.pie(class1_studs, labels= classes, explode= explode, colors = colors, autopct= "%0.1f%%", shadow =True, radius =1.4, startangle=180,textprops = textprops)
plt.show()
```

```python
wedgeprops={"linewidth":4, 'width':1, "edgecolor": "k"}
plt.pie(
         class1_studs,
         explode=explode,
         labels=classes,
         colors=colors,
         autopct="%0.2f%%",
         pctdistance=0.6,
         shadow=False,
         labeldistance=1.4,
         startangle=270,
         radius=1.1,
         counterclock=True,
         wedgeprops=wedgeprops,
         textprops=textprops,
         center=(2, 3),
         frame=True,
         rotatelabels=True
         )
plt.show()
```

```python
wedgeprops={"linewidth":4, 'width':2, "edgecolor": "k"}
plt.pie(
         class1_studs,
         explode=explode,
         labels=classes,
         colors=colors,
         autopct="%0.2f%%",
         pctdistance=0.6,
         shadow=False,
         labeldistance=1.4,
         startangle=270,
         radius=1.1,
         counterclock=True,
         wedgeprops=wedgeprops,
         textprops=textprops,
         center=(2, 3),
         frame=True,
         rotatelabels=True
         )
plt.show()
```
